````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar dalam mode grayscale
image = cv2.imread('renjun.jpg', cv2.IMREAD_GRAYSCALE)

# Hitung histogram
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Normalisasi histogram
histogram_normalized = histogram / histogram.sum()

# Visualisasi
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.title('Gambar Grayscale')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title('Histogram Asli')
plt.plot(histogram)
plt.xlabel('Intensitas Piksel')
plt.ylabel('Frekuensi')

plt.subplot(133)
plt.title('Histogram Normalisasi')
plt.plot(histogram_normalized)
plt.xlabel('Intensitas Piksel')
plt.ylabel('Probabilitas')

plt.tight_layout()
plt.show()

# Cetak informasi tambahan
print(f"Total piksel: {image.size}")
print(f"Jumlah piksel dalam histogram: {histogram.sum()}")
print(f"Jumlah probabilitas histogram normalisasi: {histogram_normalized.sum()}")
````

### **Efek Setelah Konvolusi**
Operasi konvolusi dengan kernel rata-rata menghasilkan beberapa efek berikut:

1. **Menghaluskan Gambar (Blurring)**:
   - Variasi intensitas piksel dalam gambar menjadi lebih seragam karena nilai piksel rata-rata dihitung dari tetangganya. Ini menyebabkan detail gambar menjadi lebih lembut.

2. **Mengurangi Noise**:
   - Noise (gangguan seperti piksel "acak" atau tajam) yang tidak sesuai dengan pola gambar utama menjadi teredam karena dirata-rata.

3. **Menurunkan Tajamnya Detail**:
   - Konvolusi dengan kernel rata-rata mengaburkan batas objek (contohnya tepi atau kontur). Hal ini karena nilai di tepi menjadi campuran dari piksel di sekitarnya, mengurangi kontras lokal.

**Contoh Situasi Penggunaan**:
- Mengurangi noise pada gambar sebelum deteksi fitur.
- Membuat efek blur untuk estetika gambar atau keperluan preprocessing.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar grayscale
image = cv2.imread('rem.jpg', cv2.IMREAD_GRAYSCALE)

# Transformasi Fourier
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)

# Magnitude Spectrum
magnitude_spectrum = 20 * np.log10(np.abs(fshift))

# Rekonstruksi gambar (Inverse Fourier)
f_ishift = np.fft.ifftshift(fshift)
img_back = np.abs(np.fft.ifft2(f_ishift))

# Visualisasi
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.title('Gambar Asli')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title('Magnitude Spectrum')
plt.imshow(magnitude_spectrum, cmap='jet')
plt.colorbar()

plt.subplot(133)
plt.title('Gambar Rekonstruksi')
plt.imshow(img_back, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

# Informasi tambahan
print("Interpretasi Magnitude Spectrum:\n- Menunjukkan distribusi frekuensi\n- Pusat adalah komponen DC\n- Lingkaran luar adalah frekuensi tinggi")
````

### **Informasi dari Magnitude Spectrum**
1. **Distribusi Frekuensi**:
   - **Frekuensi Rendah (pusat)**:
     - Menunjukkan komponen gambar dengan perubahan intensitas yang lambat, seperti area terang atau gelap besar.
     - Komponen ini sering berkontribusi pada struktur global gambar.
   - **Frekuensi Tinggi (pinggir)**:
     - Menunjukkan perubahan intensitas yang cepat, seperti tepi, tekstur, atau detail kecil.
     - Komponen ini sering berkontribusi pada detail gambar.

2. **Aplikasi dalam Pemrosesan Gambar**:
   - **Filter Frekuensi Rendah**: Menghaluskan gambar atau mengurangi noise.
   - **Filter Frekuensi Tinggi**: Menonjolkan tepi atau detail.
   - **Analisis Pola**: Mengevaluasi karakteristik gambar berdasarkan distribusi frekuensi.

3. **Keseimbangan Frekuensi**:
   - Gambar yang jelas dan detail memiliki distribusi frekuensi tinggi yang signifikan.
   - Gambar buram memiliki komponen frekuensi tinggi yang lebih sedikit.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
image = cv2.imread('rem(2).jpg', cv2.IMREAD_GRAYSCALE)

# Parameter pencerahan
beta = 50  # Nilai tambahan

# Pencerahan dengan clipping
image_cerah = np.clip(image.astype(np.int16) + beta, 0, 255).astype(np.uint8)

# Visualisasi
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.title('Gambar Asli')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.title(f'Gambar Cerah ( = {beta})')
plt.imshow(image_cerah, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

# Informasi tambahan
print(f"Perubahan:\n- Nilai piksel ditambah {beta}\n- Clipping mencegah overflow\n- Kontras bertambah")
````

#### **3. Parameter Pencerahan**
```python

```
- **beta**: Nilai penambahan konstan untuk semua piksel dalam gambar. Semakin besar nilai beta, semakin terang gambar.

---

#### **4. Operasi Pencerahan**
```python
image_cerah = np.clip(image.astype(np.int16) + beta, 0, 255).astype(np.uint8)
```
1. **image.astype(np.int16)**:
   - Mengonversi tipe data piksel dari `uint8` (\(0\)\(255\)) menjadi `int16` (\(-32,768\)\(32,767\)) untuk mencegah **overflow** saat menjumlahkan piksel dengan \( \beta \).
   - Jika tetap menggunakan `uint8`, nilai di atas \(255\) akan melingkar kembali ke nilai rendah (\(0\)\(255\)).

2. **+ beta**:
   - Menambahkan nilai konstan \( \beta \) (\(50\) dalam kasus ini) ke setiap piksel dalam gambar.

3. **np.clip(..., 0, 255)**:
   - Fungsi **clipping** membatasi nilai piksel agar tetap berada dalam rentang valid \(0\)\(255\).
   - Jika nilai piksel setelah penambahan melebihi \(255\), maka akan diset ke \(255\).
   - Jika nilainya kurang dari \(0\), akan diset ke \(0\).

4. **astype(np.uint8)**:
   - Mengembalikan tipe data piksel menjadi `uint8` agar dapat digunakan kembali sebagai gambar valid.

---

### **Perubahan pada Gambar**
1. **Gambar Menjadi Lebih Cerah**:
   - Setiap piksel dalam gambar menjadi lebih terang karena ditambah dengan nilai konstan \( \beta \).

2. **Peningkatan Kontras**:
   - Piksel dengan nilai tinggi menjadi lebih terang, sementara piksel dengan nilai rendah tetap rendah (atau sedikit naik).
   - Area gelap tetap lebih gelap dibandingkan area terang, menghasilkan kesan peningkatan kontras visual.

3. **Efek Clipping**:
   - Piksel dengan nilai tinggi mendekati \(255\) akan mencapai nilai maksimum, sehingga dapat mengurangi detail pada area sangat terang.
